```python
import selenium
import pandas as pd
import re

from selenium import webdriver 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from selenium.webdriver.support import expected_conditions as EC

import warnings
warnings.filterwarnings('ignore')
import time
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.olx.in")
```

```python
all_cat = driver.find_element(By.XPATH,'/html/body/div/div/div[1]/div/div/div[1]/div[1]/span[2]/button')
all_cat.click()
```

```python
for_sale = driver.find_element(By.XPATH,'/html/body/div/div/div[1]/div/div/div[2]/div/div[2]/div[2]/a')
for_sale.click()
```

```python
product_url = []
for pages in range(0,2):
    time.sleep(20)
    url = driver.find_elements(By.XPATH,'//li[@data-aut-id="itemBox"]/a')
    for i in url:
        product_url.append(i.get_attribute('href'))
    load_more = driver.find_element(By.XPATH,'//div[@class="_38O09"]')
    load_more.click()
```

```python
product_url=list(set(product_url))
```

```python
product_urls =  pd.DataFrame(product_url, columns=['LINK'])
```

```python
product_urls.to_csv('URLS.csv',index=False)
```

```python

d = {"Property type":[], "Bedroom":[], "Bathroom":[], "Furnished Status":[],"Carpet Area":[],"Construction Status":[], "Floor No":[], "SquareFeet":[], "Facing":[],"Car Parking":[],"Total Price (INR)":[],"Address":[],"BHK":[]}
for i in product_urls['LINK'][0:20]:
    driver.get(i)
    time.sleep(2)
    
    try:
        brand_element = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_type"]')))
        d["Property type"].append(brand_element.text)
    except:
        d["Property type"].append("-")
    
    try:
        product_name_element = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_rooms"]')))
        d["Bedroom"].append(product_name_element.text)
    except:
        d["Bedroom"].append("-")
        
    try:
        price = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_bathrooms"]')))
        d["Bathroom"].append(price.text)
    except:
        d["Bathroom"].append("-")
    
    try:
        returnexc = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_furnished"]')))
        d["Furnished Status"].append(returnexc.text)
    except:
        d["Furnished Status"].append("-")
        
    try:
        dele = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_carpetarea"]')))
        d["Carpet Area"].append(dele.text)
    except:
        d["Carpet Area"].append("-")
    
    try:
        
        ava = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_constructionstatus"]')))
        d["Construction Status"].append(ava.text)

    except:
        d["Construction Status"].append("-")

    try:
        brand_element = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_totalfloors"]')))
        d["Floor No"].append(brand_element.text)
    except:
        d["Floor No"].append("-")
    
    try:
        product_name_element = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_ft"]')))
        d["SquareFeet"].append(product_name_element.text)
    except:
        d["SquareFeet"].append("-")
        
    try:
        price = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_facing"]')))
        d["Facing"].append(price.text)
    except:
        d["Facing"].append("-")
    
    try:
        returnexc = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="value_carparking"]')))
        d["Car Parking"].append(returnexc.text)
    except:
        d["Car Parking"].append("-")
        
    try:
        dele = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@data-aut-id="itemPrice"]')))
        #cleaned_text = re.sub(r'[₹,]', '', dele).strip()
        d["Total Price (INR)"].append(dele.text)
    except:
        d["Total Price (INR)"].append("-")
            
    try:
        dele = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//div[@data-aut-id="itemLocation"]')))
        loc=[]
        #loc.append(dele.text)        
        #omponents = [component.strip() for component in location.split(',')]
        d["Address"].append(dele.text)
    except:
        d["Address"].append("-")

    try:
        dele = WebDriverWait(driver, 0.1).until(EC.visibility_of_element_located((By.XPATH, '//span[@class="dBLgK"]')))
        #match = re.search(r'\d', dele)

        #if match:
         #   first_digit = match.group()
     
        d["BHK"].append(dele.text)
    except:
        d["BHK"].append("-")
        
```
